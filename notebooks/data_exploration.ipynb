```python

def scale_data(observations, method='standard'):

    scaled_observations = observations.copy()

    if method == 'standard':

        scaler = StandardScaler()

    elif method == 'robust':

        scaler = RobustScaler()
    scaled_observations['VALUE_NUMERIC'] = scaler.fit_transform(
    scaled_observations[['VALUE_NUMERIC']]
    )   
    return scaled_observations 
```

```python
def apply_mode_imputation(observations):

    processed_observations = observations.copy()

    missing_cols = ['ENCOUNTER','CATEGORY','UNITS']

    mode_imputer = SimpleImputer(strategy='most_frequent')

    processed_observations[missing_cols] = mode_imputer.fit_transform(
        processed_observations[missing_cols]
    )

    return processed_observations
```

```python
def apply_constant_imputation(observations):

    processed_observations = observations.copy()

    missing_cols = ['ENCOUNTER','CATEGORY','UNITS']

    constant_imputer = SimpleImputer(
        strategy='constant',
        fill_value='Unknown'
    )

    processed_observations[missing_cols] = constant_imputer.fit_transform(
        processed_observations[missing_cols]
    )

    return processed_observations
```

```python
def apply_keep_outliers(observations):

    return observations.copy()
```

```python
def apply_iqr_outliers(observations):

    processed_observations = observations.copy()

    col = 'VALUE_NUMERIC'

    Q1 = processed_observations[col].quantile(0.25)

    Q3 = processed_observations[col].quantile(0.75)

    IQR = Q3-Q1

    lower = Q1-1.5*IQR

    upper = Q3+1.5*IQR

    processed_observations = processed_observations[
        (processed_observations[col]>=lower)
        &
        (processed_observations[col]<=upper)
    ]

    return processed_observations
```

```python
def create_numeric_column(observations):

    processed_observations = observations.copy()

    processed_observations['VALUE_NUMERIC'] = pd.to_numeric(

        processed_observations['VALUE'],

        errors='coerce'

    )

    processed_observations.dropna(

        subset=['VALUE_NUMERIC'],

        inplace=True

    )

    return processed_observations
```

```python
def apply_pipeline(observations,pipeline):

    processed_observations = observations.copy()
    if pipeline['missing_method']=='mode':

        processed_observations = apply_mode_imputation(
            processed_observations
        )

    elif pipeline['missing_method']=='constant':

        processed_observations = apply_constant_imputation(
            processed_observations
        )

        #outlier engine

    if pipeline['outlier_method']=='keep':

        processed_observations = apply_keep_outliers(
            processed_observations
        )

    elif pipeline['outlier_method']=='iqr':

        processed_observations = apply_iqr_outliers(
            processed_observations
        ) 

    processed_observations = create_numeric_column(
    processed_observations
)   
    print(processed_observations.columns)


    #scalling Engine
    processed_observations = scale_data(

        processed_observations,

        method=pipeline['scaler']

    )
    
#return  Final DataSet
    return processed_observations
```

```python

# Pipeline Definitions

pipeline1 = {

    'missing_method': 'mode',

    'outlier_method': 'keep',

    'scaler': 'standard'

}

pipeline2 = {

    'missing_method': 'mode',

    'outlier_method': 'keep',

    'scaler': 'robust'
    
}
```

```python
#Pipeline Evaluation Engine
def evaluate_pipeline(observations):
    score = {}
    score ['rows'] = observations.shape[0]
    score ['columns'] = observations.shape[1]

    score['missing_values'] = observations.isnull().sum().sum()

    score['duplicates'] = observations.duplicated().sum()

    return score
```
